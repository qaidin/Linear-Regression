# PROSES PREPROCESSING

**Drop feature 'ID' karena feature ini tidak diperlukan**

````
df.drop('id',axis=1,inplace=True)
````

**mengubah nama kolom menjadi lowercase untuk mempermudah melakukan cleaning data**

````
df.columns = df.columns.str.lower()
````


**Drop feature**
```
# ['poolqc','fence','miscfeature','fireplacequ','lotfrontage']
```

karena memiliki terlalu banyak nilai NaN(None a Number) kita juga tidak bisa mengisi seluuruh data yang kosong dengan metode .mean() ataupun metode missing value lainnya.. 


````
df.drop(['poolqc','fence','miscfeature','fireplacequ','lotfrontage'],axis=1, inplace=True)
````

**Identifikasi featuures yang memiliki nilai NaN lainnya**

> bisa kita lihat data dibawah memiliki kolom 'alley' memiliiki 1369 data kosong yang dimana ituu adalah data kosong yang sangat banyak. lebih baik lakuukan metode drop([df.alley])



````
df.isna().sum()
````

````
df.drop('alley',axis=1,inplace=True)
````

**setelah melakkan proses preproocessing didapatkan data yang cukup bagus dengan 1338 entries dan 74 kolom. hal selanjutnya adalah mengubah semua type data menjadi int64 menggunakan library inbuilt sklearn 'LabelEncooder()' **

````
df.dropna(inplace=True, axis=0)
df.info()
````

**pemisahan data independent dan data dependent**

> y = df['saleprice'] adalah target data yang tidak perluu di encoder lagi karena meruupakan titik acuuan dari model prediksi yang kita buat

> `x = df.iloc[:,:74]`adalah independent data yang memiliki muultivariant type data yang kita ingin seragamkan menjadi int64 menggunakan LabelEncoder()





````
x = df.iloc[:,:74]
y = df['saleprice']
````

**Proses Label Encoder**



> semuua type data menjadi int64



````
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x = x.apply(LabelEncoder().fit_transform)
x.info()
````

**Buuilt Machine Learning Proses by Sklearn library**

````

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
````

**Lakukan split data untuuk membedakan data train dan data test**

````
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
````

inisiasi variabel fungsi linear model, prediksi linear

````
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)
ypred=lm.predict(xtest)
print('koefisien : ', lm.coef_)
df.corr(method='kendall')
````

**Proses Evaluuasi Model MSE, AMSE, R2 Score**

````
from sklearn import metrics
print("Mean squared error: %.2f"
      % mean_squared_error(ytest, ypred))
````

````
absolut = metrics.mean_absolute_error(ytest, ypred)
absolut
````

````
print('Variance score: %.2f' % r2_score(ytest,ypred))
````
